```python
import pandas as pd
import numpy as np

#to reduce the training set, since it is too large
```

```python
#load the data
trainingset = pd.read_csv('/data/aai/scratch/jchan/denoise/PAUS/output_save/modify_pn2v/selectdata.csv')
stamp_train = np.load('/data/aai/scratch/jchan/denoise/PAUS/output_save/modify_pn2v/stamp_train.npy')
mask_train = np.load('/data/aai/scratch/jchan/denoise/PAUS/output_save/modify_pn2v/mask_train.npy')

mask_gal = np.load('/data/aai/scratch/jchan/denoise/PAUS/output_save/modify_pn2v/mask_gal.npy')
mask_annulus = np.load('/data/aai/scratch/jchan/denoise/PAUS/output_save/modify_pn2v/mask_annulus.npy')
area_annulus = np.load('/data/aai/scratch/jchan/denoise/PAUS/output_save/modify_pn2v/area_annulus.npy')
```

```python
trainingset
```

```python
sorted_trainingset = trainingset.sort_values('ref_id')
sorted_trainingset = sorted_trainingset.reset_index(drop=True)
sorted_trainingset
```

```python
original_indices = sorted_trainingset['Unnamed: 0'].values

shuffled_stamp_train = stamp_train[original_indices]
shuffled_mask_train = mask_train[original_indices]
shuffled_mask_gal = mask_gal[original_indices]
shuffled_mask_annulus = mask_annulus[original_indices]
shuffled_area_annulus = area_annulus[original_indices]
```

```python
first_20000 = trainingset.iloc[:20000]

shuffled_stamp_train = shuffled_stamp_train[:20000]
shuffled_mask_train = shuffled_mask_train[:20000]
shuffled_mask_gal = shuffled_mask_gal[:20000]
shuffled_mask_annulus = shuffled_mask_annulus[:20000]
shuffled_area_annulus = shuffled_area_annulus[:20000]

np.save('/data/aai/scratch/jchan/denoise/PAUS/output_save/modify_pn2v/mask_gal_reduce.npy', shuffled_mask_gal)
np.save('/data/aai/scratch/jchan/denoise/PAUS/output_save/modify_pn2v/mask_annulus_reduce.npy', shuffled_mask_annulus)
np.save('/data/aai/scratch/jchan/denoise/PAUS/output_save/modify_pn2v/area_annulus_reduce.npy', shuffled_area_annulus)
np.save('/data/aai/scratch/jchan/denoise/PAUS/output_save/modify_pn2v/stamp_train_reduce.npy', shuffled_stamp_train)
np.save('/data/aai/scratch/jchan/denoise/PAUS/output_save/modify_pn2v/mask_train_reduce.npy', shuffled_mask_train)
first_20000.to_csv('/data/aai/scratch/jchan/denoise/PAUS/output_save/modify_pn2v/selectdata_reduce.csv')
```
